## Stock Price Prediction

#### Dataset

```python
import pandas as pd

dataframe = pd.read_csv("google.csv")
dataframe.head()
```






#### Dropping the Data Attribute

```python
dataframe.drop('Date', axis=1, inplace=True)
dataframe
```

#### Data Normalization  

```python
from sklearn.preprocessing import Normalizer, MinMaxScaler

column_name = dataframe.columns
data = Normalizer().fit_transform(dataframe.transpose())
normalized_data = pd.DataFrame(data.transpose(), columns=column_name)
normalized_data

# columns = ['Open', 'Close', 'Volume']
# dataframe[columns] = Normalizer().fit_transform(dataframe[columns])
```

```python
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (18, 12)

fig, ax = plt.subplots()
ax.plot(range(len(dataframe), 0,-1), normalized_data['Volume'], 'Blue')

ax.format_ydata = 'price'
ax.set_title('Google Trading')

# Add labels
plt.ylabel('Price USD')
plt.xlabel('Trading Days')

plt.show()
```

#### Polynomial Regression using one feature





















































```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

#### creating trainX and trainY

```python
X = pd.DataFrame(normalized_data, columns=['Open','Low','High','Volume'])
y = pd.DataFrame(normalized_data, columns=['Close'])
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
print(X_train.shape,X_test.shape, y_train.shape, y_test.shape)
```

```python
poly = PolynomialFeatures(2)
poly_X = poly.fit_transform(X_train)
```

```python
model = LinearRegression()
model.fit(poly_X,y_train)
```

```python
poly_test = poly.fit_transform(X_test)
```

```python
y_pred = model.predict(poly_test)
```

```python
y_test.shape
```

```python
#print(model.score(y_pred,y_test))
from sklearn.metrics import mean_squared_error
accuracy = mean_squared_error(y_test,y_pred)
```

```python
print(accuracy*100)
```

```python
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (18, 12)

fig, ax = plt.subplots()
ax.plot(range(len(y_test), 0,-1), y_test, 'Blue')
fig, ax = plt.subplots()
ax.plot(range(len(y_pred), 0,-1), y_pred, 'Yellow')

ax.format_ydata = 'price'
ax.set_title('Google Trading')

# Add labels
plt.ylabel('Price USD')
plt.xlabel('Trading Days')

plt.show()
```
